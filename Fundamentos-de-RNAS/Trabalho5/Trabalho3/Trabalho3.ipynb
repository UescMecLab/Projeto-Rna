### Problema de ajuste de curvas por dados

Nas aplicações de ajuste de curvas, buscamos modelar uma função $f: \mathbb{R}^{p}\rightarrow \mathbb{R}^{l}$ que relacione variáveis de entrada  $x  \in \mathbb{R}^{p}$ e saídas desejadas $y  \in \mathbb{R}^{l}$. Dados os pares no conjunto de dados $D= \{(x_{1},y_{1}), \ldots, (x_{n},y_{n})\}$,  o objetivo é encontrar uma função aproximada  $h: \mathbb{R}^{p} \rightarrow \mathbb{R}^{l}$ que mapeie $x$ para $\hat{Y}$, proporcionando um ajuste eficaz que forneça uma generalização para além dos dados de treinamento. Nesse contexto, uma abordagem que parta dos passos a seguir pode facilitar o processo de obtenção de $h$.

Primordialmente convém obter os dados relevantes que descrevem a função de ajuste de curva.




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# @brief: importa dados do excel
def GetData():
  df = pd.read_excel("./Trabalho3dados.xlsx")
  x = df['x']
  y = df['y']
  z = df['z']
  return(np.array([x, y, z]))
```

### 1. Avaliar os dados


A etapa de avaliação de dados é crucial  para garantir que os dados sejam adequados e confiáveis para o ajuste do modelo. Nesse estágio, convém analisar se os dados foram adequadamente medidos além de avaliar se eles catacterizam a função em termos de quantidade e abrangência.

#### **Normalização**

A normalização dos dados é um processo utilizado para ajustar os valores das variáveis em uma escala comum, a fim de evitar que a grandeza de algumas variáveis domine ou influencie excessivamente o algoritmo, foram escolhidos dois métodos de normalização:

Algumas abordagens distintas garantem  que as variáveis estão na mesma escala.

- **ZScore:**
    A normalização por Escore Z é empregada para transformar os dados de forma que sua média seja zero e seu desvio padrão seja igual a um.  Isso é feito subtraindo a média dos dados da variável e dividindo pelo desvio padrão
$$X^i_{norm} = \frac{X^i - \mu}{\sigma} $$




```python
def ZScore(Data):
  return ((Data - np.mean(Data)) / np.std(Data))
```

 - **Mínimos e Máximos:** É possível normalizar a dimensão $i$ do conjunto de dados $X$ entre $N_{min}$ e $N_{max}$ de modo que eles se enquadrem no intervalo especificado (por exemplo $[0, 1]$), ou seja, onde o valor mínimo original ( min) é mapeado para 0 e o valor máximo original ( max) é mapeado para 1, ajustadndo linearmente e de forma proporcional os valores, mantendo suas relações e preservando a distribuição original.

 $$X_{norm}^{i} = Nmin+ \frac{(X^{i}-Xmin)(Nmax-Nmin)}{Xmax-Xmin}$$

Essa normalização também coloca todos os dados em uma escala expressiva e é útil quando você quer garantir que todos os dados estejam dentro de um intervalo específico.

```python
def MinMax(Data, min_value=-1, max_value=1):
    normalized_data = min_value + (Data - np.min(Data)) * (max_value - min_value) / (np.max(Data) - np.min(Data))
    return normalized_data
```

- **Vetor unitário:** Aqui  os vetores são ajustados para que tenham comprimento igual a um, e mantenham a direção e o ângulo. Essa abordagem é útil em situações onde a magnitude dos vetores não é relevante, mas sim suas direções relativas.
$$X^i_{norm} = \frac{X^i}{||X^i||} $$

```python
def UnitVectorNormalization(Data):
    norms = np.linalg.norm(Data)
    normalized_data = Data / norms
    return normalized_data
```

```python
# @brief: normaliza n-1 variaveis em x
def GetNormData(method=ZScore):
  x = GetData()
  for i in range(len(x)):
    x[i] = method(x[i])
  return (x)

# @brief: plota os dados importados
def PlotData(fig):
  x, y, z = GetData()
  ax = fig.add_subplot(1, 4, 1, projection='3d')
  ax.scatter(x, y, z, c='blue', marker='o')
  ConfGraph3D(ax, 'x', 'y', 'z', "Dados importados")
  return (ax)

def PlotNormalizedData(fig, method, pos, label, color):
  xNorm, yNorm, zNorm = GetNormData(method)
  ax = fig.add_subplot(1, 4, pos, projection='3d')
  ax.scatter(xNorm, yNorm, zNorm, c=color, marker='o')
  ConfGraph3D(ax, 'x', 'y', 'z', "Dados Normalizados por " + label)

def ConfGraph3D(ax, x, y, z, title):
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    ax.set_zlabel(z)
    ax.set_title(title)

def PlotFigure():
  fig = plt.figure(figsize=(16, 8))
  PlotData(fig)
  PlotNormalizedData(fig, ZScore, 2, 'zscore', 'red')
  PlotNormalizedData(fig, MinMax, 3, 'minmax', 'green')
  PlotNormalizedData(fig, UnitVectorNormalization, 4, 'unit vetor', 'orange')

PlotFigure()
```

Aplicando os diferentes metódos de normalização, os dados são preparados de maneira adequada para a aplicação de algoritmos.

```python
def DenormalizeData(Data, normMethod, Orig):
  if (normMethod == MinMax):
    return (Data * (np.max(Orig) - np.min(Orig)) / 2 + (np.max(Orig) + np.min(Orig)) / 2)
  return (Data * Orig.std() + Orig.mean())

def PlotDenormalizedData(fig, method, pos, label, color):
  xNorm, yNorm, zNorm = GetNormData(method)
  x, y, z = GetData()
  xDenorm = DenormalizeData(xNorm, method, x)
  yDenorm = DenormalizeData(yNorm, method, y)
  zDenorm = DenormalizeData(zNorm, method, z)
  ax = fig.add_subplot(1, 3, pos, projection='3d')
  ax.scatter(xDenorm, yDenorm, zDenorm, c=color, marker='o')
  ConfGraph3D(ax, 'x', 'y', 'z', "Dados desnormalizados por " + label)
  return (xNorm, yNorm)

def PlotDenorm():
  fig = plt.figure(figsize=(16, 8))
  PlotDenormalizedData(fig, ZScore, 2, 'zscore', 'red')
  PlotDenormalizedData(fig, MinMax, 3, 'minmax', 'green')

PlotDenorm()
```

### 2. Propor uma estrutura para $h$:

A ideia é encontrar um modelo que seja capaz de capturar as características importantes dos dados de treinamento de forma eficiente e, ao mesmo tempo, seja simples o suficiente para evitar overfitting (ajuste excessivo aos dados de treinamento) e facilitar a interpretação dos resultados.

 > O melhor modelo é o mais simples que permite avaliar o fenômeno estudado.

```python
def CurveFitFunction(p, x):
    return ((p[0] * x[0]**3) + (p[1] * x[1]**2) + p[2])
```

### 3. Definir uma medida de erro $J(e)$:

Dada a restrição, encontrar $a,b,c$ que ajustem a curva  $z = ax^3 + by^2 + c$ ao conjunto de dados, podemos tornar o problema irrestrito ao minimizar o erro da estrutura proposta pelos parâmetros encontrados.

O objetivo aqui, é quantificar o quanto o modelo $h$ se desvia das respostas reais $Y_{i}$. No ajuste de curva, busca-se minimizar o erro entre as respostas reais ($Y_{i}$) e as respostas preditas ($\hat{Y}_{i}=h(x_{i})$) , $e_{i}=Y_{i}-\hat{Y}_{i}$. Para isso, precisa-se escolher uma medida de erro $J(e)$ com mínimo definido.

- **Erro Quadrático Médio (MSE):** =$\frac{1}{n}\sum_{i=1}^{n} e_{i}^{2}$

  - É a média dos quadrados dos erros entre as respostas reais e as respostas preditas pelo modelo.
  - O MSE é amplamente utilizado por ser fácil de derivar e otimizar, mas pode ser sensível a outliers.

- **Raiz do Erro Quadrático Médio (RMSE):**=$\sqrt{ MSE}$

  - É utilizada para tornar a medida de erro mais interpretável, comumente usada para apresentar o erro na mesma escala das unidades das variáveis de resposta.
  - Por usar a raiz do erro quadrático médio, reduz a influência do quadrado na interpretação da magnitude do erro

- **Erro Absoluto Médio (MAE):**=$\frac{1}{n}\sum_{i=1}^{n} \left |  e_{i}\right |$
  - É a média dos valores absolutos dos erros  e as respostas preditas pelo modelo

  - é menos sensível a outliers do que o MSE, mas pode ter dificuldades computacionais na otimização.
  - Precisamos obter o gradiente da medida de erro, busca-se funções de derivação mais simples

- **Erro Percentual Absoluto Médio (MAPE)**:=$\frac{100}{n}\sum_{i=1}^{n} \left |  \frac{Yi-\hat{Y}i}{Yi}\right |$
  - É uma medida de erro relativa que expressa o erro em termos de uma porcentagem da resposta real
  - Essa medida é útil quando as variáveis de resposta têm magnitudes muito diferentes, pois o erro é normalizado em relação ao valor real.
  - Por ser relativo (dividir por $Y_{i}$), depende menos da magnitude de $Y_{i}$.


```python
def MSE(zPred, zReal):
    return (np.mean(zReal - zPred)**2)

def RMSE(zPred, zReal):
    return (np.sqrt(MSE(zPred, zReal)))

def MAE(zPred, zReal):
    return (np.mean(np.abs(zReal - zPred)))

def MAPE(zPred, zReal):
    return (np.mean(np.abs((zReal - zPred) / zReal)) * 100)
```

### 4. Aplicar minimização para obter os parâmetros do modelo.

Definida a medida de erro, é possível tornar o problema restrito irrestrito considerando a minimização da função de custo. Em um contexto de ajuste de curvas, a função de custo geralmente mede uma discrepância entre os valores reais ($Z_i$) e os valores previstos pelo modelo ($\hat{Z_i}$) para um conjunto de dados.

- Dado o conjunto de dados D={$(x_{1},y_{1}),(x_{2},y_{2}), ..., (x_{n},y_{n})$}, uma estrutura $\hat{Z}=h(x)$ para o modelo com conjunto de parâmetros $W$ e uma medida de erro $E(e)$, com $e_{i}=Z_{i}-\hat{Z}_{i}$, tem-se :

$$
min_{w} E(e) ; \hat{Z}=h(x)
$$

- Como $e=Z-\hat{Z}$, transforma-se o problema de otimização com restrição em um problema irrestrito

$$
 min_{w} E(Z-h(x))
 $$


Na implementação do algoritmo para a  minimização da função de custo, o gradiente é uma ferramenta essencial pois aponta na direção do aumento maior da função de erro (ou o oposto, se estiver minimizando).

**Calculando o Gradiente das funções de erro:**

> Relembrando a regra da cadeia:
 $$\frac{\partial f(g(x))}{\partial x}= f'(g(x))*g'(x)$$

- **Erro Quadrático Médio (MSE):** =$\frac{1}{n}\sum_{i=1}^{n} e_{i}^{2}$

$$MSE = \frac{1}{n} \sum_{i=1}^{n} (Z_i - \hat{Z}_i)^2 =  \frac{1}{n} \sum_{i=1}^{n} (Z_i - ax_i^3 - by_i^2-c)^2$$
$$\Rightarrow \nabla MSE = -\frac{2}{n}\sum_{i=1}^{n}((Z_i - \hat{Z}_i)[x_i^3, y_i^2, 1]) $$

- **Raiz do Erro Quadrático Médio (RMSE):**=$\sqrt{ MSE}$
$$\sqrt{MSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} (Z_i - \hat{Z}_i)^2} = \frac{1}{\sqrt{n}}\sum_{i=1}^{n}  \sqrt{(Z_i - ax_i^3 - by_i^2-c)^2}$$

$$\Rightarrow \nabla \sqrt{MSE} = -\frac{1}{\sqrt{n}} \sum_{i=1}^{n} ((Z_i - \hat{Z}_i)[x_i^3, y_i^2, 1] )$$

- **Erro Absoluto Médio (MAE):**=$\frac{1}{n}\sum_{i=1}^{n} \left |  e_{i}\right |$

$$ MAE = \frac{1}{n}\sum_{i=1}^{n} \left | (Z_i - \hat{Z}_i) \right | = \frac{1}{n}\sum_{i=1}^{n} \left | (Z_i - ax_i^3 - by_i^2-c) \right | $$

$$\Rightarrow \nabla {MAE} = -\frac{1}{n} \sum_{i=1}^{n}\frac{Z_i - \hat{Z_i}}{|Z_i - \hat{Z_i}|}[x_i^3, y_i^2, 1] $$

- **Erro Percentual Absoluto Médio (MAPE)**:=$\frac{100}{n}\sum_{i=1}^{n} \left |  \frac{Zi-\hat{Z}i}{Zi}\right |$

$$ MAPE = \frac{100}{n}\sum_{i=1}^{n} \left |  \frac{Zi-\hat{Z}i}{Zi}\right|= \frac{100}{n}\sum_{i=1}^{n} \left |  \frac{Zi - ax_i^3 - by_i^2-c}{Zi}\right| $$

$$\Rightarrow \nabla {MAPE} = -\frac{100}{n}\sum_{i=1}^{n}  \frac{|Z_i|}{Z_i^2} \frac{Z_i -  \hat{Z_i}}{|Z_i -  \hat{Z_i}|} [x_i^3, y_i^2, 1]$$

```python
def GradMSE(x, zReal, zPred, n):
    errors = (zPred - zReal)
    GradA = 2/n * np.sum((x[0] ** 3) * errors)
    GradB = 2/n * np.sum((x[1] ** 2) * errors)
    GradC = 2/n * np.sum((1) * errors)
    return [GradA, GradB, GradC]

def GradRMSE(x, zReal, zPred, n):
    errors = (zPred - zReal)
    GradA = 1/np.sqrt(n) * np.sum((x[0] ** 3) * errors)
    GradB = 1/np.sqrt(n) * np.sum((x[1] ** 2) * errors)
    GradC = 1/np.sqrt(n) * np.sum((1) * errors)
    return [GradA, GradB, GradC]

def GradMAE(x, zReal, zPred, n):
    errors = (zPred - zReal)
    errorsNorm = np.abs(errors)
    GradA = 1/n * np.sum((x[0] ** 3) * errors / errorsNorm)
    GradB = 1/n * np.sum((x[1] ** 2) * errors / errorsNorm)
    GradC = 1/n * np.sum(errors / errorsNorm)
    return [GradA, GradB, GradC]

def GradMAPE(x, zReal, zPred, n):
    errors = np.sign((zPred - zReal))
    errorsNorm = np.abs(errors)
    zRealNorm = np.abs(zReal)
    GradA = 100/n * np.sum((x[0] ** 3) * (errors / errorsNorm) * (zRealNorm / zReal ** 2))
    GradB = 100/n * np.sum((x[1] ** 2) * (errors / errorsNorm) * (zRealNorm / zReal ** 2))
    GradC = 100/n * np.sum(((errors / errorsNorm) * (zRealNorm / zReal ** 2)))
    return [GradA, GradB, GradC]
```

Em sintese, o gradiente de uma função de custo é uma ferramenta importante para otimizar modelos pois guia a busca pelos melhores valores dos parâmetros, levando a uma convergência mais rápida e eficaz.

```python
def CheckCost(HistCost, tol=0.0001):
  if (len(HistCost) <= 2):
    return False
  return (abs(HistCost[-1] - HistCost[-2]) < tol)

def GradientDescent(p, K, GradCostFunc, CostFunc, normMethod, alpha):
    HistCost =  []
    DataNorm = GetNormData(normMethod)
    zReal = DataNorm[2]
    while (CheckCost(HistCost) == False):
        zPred = CurveFitFunction(p, DataNorm)
        HistCost.append(CostFunc(zPred, zReal))
        Grad = GradCostFunc(DataNorm, zReal, zPred, len(DataNorm[0]))
        p = p - np.dot(Grad, alpha)
        K -= 1
        if (K == 0):
          break
    return (p, HistCost)
```

Assim com o objetivo de minimizar a função de custo, o algoritmo do gradiente descendente permite o ajuste dos parâmetros iterativamente em direção ao mínimo local ou global.

```python
def GetZDenorm(p, normMethod, ZOrig):
  x = GetNormData(normMethod)
  X, Y = np.meshgrid(np.linspace(min(x[0]), max(x[0]), 100),
                     np.linspace(min(x[1]), max(x[1]), 100))
  ZNorm = (CurveFitFunction(p, [X, Y]))
  ZDenorm = DenormalizeData(CurveFitFunction(p, [X, Y]), normMethod, ZOrig)
  return (ZDenorm)

def GetSurface(x, GradCostFunc, CostFunc, k, normMethod, alpha):
  p, HistCost = GradientDescent([1.0, 1.0, 1.0], k, GradCostFunc, CostFunc, normMethod, alpha)
  X, Y = np.meshgrid(np.linspace(min(x[0]), max(x[0]), 100),
                     np.linspace(min(x[1]), max(x[1]), 100))
  Z = GetZDenorm(p, normMethod, x[2])
  return (X, Y, Z, HistCost)

def PlotSurface(fig, X, Y, Z, label):
  ax = fig.add_subplot(1, 2, 1, projection='3d')
  ax.plot_surface(X, Y, Z, alpha=0.8, cmap='viridis')
  ConfGraph3D(ax, 'x', 'y', 'z', "Ajuste de curva por " + label)
  return (ax)

def PlotCosts(fig, Hist, legenda):
  ax = fig.add_subplot(1, 2, 2)
  k = np.linspace(0, len(Hist), len(Hist))
  ax.plot(k, Hist)
  ax.set_xlabel('Iterações (k)')
  ax.set_ylabel('Custo')
  ax.set_title("Histórico de custo da função " + legenda)
  ax.grid(True)

def TestCostFunc(GradCostFunc, CostFunc, label, normMethod=ZScore, K=1000, alpha=0.001):
  x = GetData()
  fig = plt.figure(figsize=(12, 6))
  X, Y, Z, HistCost = GetSurface(x, GradCostFunc, CostFunc, K, normMethod, alpha)
  ax = PlotSurface(fig, X, Y, Z, label)
  ax.scatter(x[0], x[1], x[2], c='blue', marker='o')
  PlotCosts(fig, HistCost, label)

```
